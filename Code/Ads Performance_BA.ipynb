```python
import pandas as pd
import os
import sys
import re

import gspread
from gspread_dataframe import set_with_dataframe
from oauth2client.service_account import ServiceAccountCredentials
from google.oauth2.service_account import Credentials
```

```python
# Đường dẫn thư mục chứa các file CSV
folder_path = "US/BA_US/2024"

# Danh sách các file CSV
file_names = [
    "US_Quarter 1_2024.csv",
    "US_Quarter 2_2024.csv",
    "US_Quarter 3_2024.csv",
    "US_Quarter 4_2024.csv"
]

# Danh sách để chứa các dataframe sau khi đọc
dfs = []

for file in file_names:
    file_path = os.path.join(folder_path, file)
    df = pd.read_csv(file_path, skiprows=1)

    # Chuyển Reporting Date sang datetime
    df["Reporting Date"] = pd.to_datetime(df["Reporting Date"], errors="coerce")

    # Tính quarter từ Reporting Date
    df["quarter"] = df["Reporting Date"].dt.quarter

    dfs.append(df)

# Gộp tất cả lại
merged_df = pd.concat(dfs, ignore_index=True)
```

```python
import pandas as pd
import os
import sys
import re

import gspread
from gspread_dataframe import set_with_dataframe
from oauth2client.service_account import ServiceAccountCredentials
from google.oauth2.service_account import Credentials
```

```python
# ASIN Title, ASIN, Category, Impressions: Impressions, Clicks: Clicks, Clicks: Click Rate (CTR), Cart Adds: Cart Adds, Purchases: Purchases, Purchases: Search Traffic Sales, Purchases: Conversion Rate %, Reporting Date

# Thư mục chứa dữ liệu
folder_path = "/Users/ppcintern/Downloads/ADS Performance/US/BA_US/2025"

# Tên file các tháng 4, 5, 6
monthly_files = [
    "US_Month 4_2025.csv",
    "US_Month 5_2025.csv",
    "US_Month 6_2025.csv"
]

# Các cột cần lấy
selected_columns = [
    "ASIN", "Category",
    "Impressions: Impressions", "Clicks: Clicks", "Clicks: Click Rate (CTR)",
    "Cart Adds: Cart Adds", "Purchases: Purchases",
    "Purchases: Search Traffic Sales", "Purchases: Conversion Rate %"
]

# Đọc và lọc dữ liệu từng tháng
dfs = []
for file in monthly_files:
    path = os.path.join(folder_path, file)
    df = pd.read_csv(path, skiprows=1)  # Nếu có dòng đầu để bỏ
    df = df[selected_columns]
    dfs.append(df)

# Gộp 3 tháng
merged_months = pd.concat(dfs, ignore_index=True)

# Tính toán: groupby và tính toán tương ứng
agg_dict = {
    "Impressions: Impressions": "sum",
    "Clicks: Clicks": "sum",
    "Clicks: Click Rate (CTR)": "mean",
    "Cart Adds: Cart Adds": "sum",
    "Purchases: Purchases": "sum",
    "Purchases: Search Traffic Sales": "sum",
    "Purchases: Conversion Rate %": "mean"
}

monthly_summary = merged_months.groupby(["ASIN", "Category"], as_index=False).agg(agg_dict)

# Gán ngày cuối quý 2
monthly_summary["Reporting Date"] = pd.to_datetime("2025-06-30")

# Đọc dữ liệu quý 1
q1_file = os.path.join(folder_path, "US_Quarter 1_2025.csv")
q1_df = pd.read_csv(q1_file, skiprows=1)
q1_df = q1_df[selected_columns]
q1_df["Reporting Date"] = pd.to_datetime("2025-03-31")

# Gộp dữ liệu Q1 và tổng hợp tháng 4-6
final_df = pd.concat([q1_df, monthly_summary], ignore_index=True)

# Convert Reporting Date và tạo cột quarter từ đó
final_df["Reporting Date"] = pd.to_datetime(final_df["Reporting Date"])
final_df["Quarter"] = final_df["Reporting Date"].dt.quarter

# Hiển thị kết quả
final_df.tail()
```
