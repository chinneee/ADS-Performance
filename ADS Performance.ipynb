# US MARKET

```python
import pandas as pd

import gspread
from gspread_dataframe import set_with_dataframe
from oauth2client.service_account import ServiceAccountCredentials
from google.oauth2.service_account import Credentials
```

```python
data_us = pd.read_csv('/Users/ppcintern/Downloads/ADS Performance/US_All_Listing.txt', sep='\t')
# remove " PDT" from 'open-date' 
data_us['open-date'] = data_us['open-date'].str.replace(' PDT', '', regex=False)

# Xoá phần " PST" hoặc bất kỳ múi giờ nào phía sau
data_us['open-date'] = data_us['open-date'].str.replace(r'\s+[A-Z]+$', '', regex=True)

# Parse lại với format cụ thể
data_us['open-date'] = pd.to_datetime(data_us['open-date'], format='%Y-%m-%d %H:%M:%S')

# Select all record with the oldest open-date
oldest_open_date_us = data_us['open-date'].min()
data_us_oldest = data_us[data_us['open-date'] == oldest_open_date_us]
data_us_oldest
```

```python
data_us.info()
```

```python
# Lấy những cột cần thiết: asin1, item-name, listing-id, seller-sku, price, open-date, product-id, fulfillment-channel, merchant-shipping-group, status
data_us = data_us[['asin1', 'item-name', 'listing-id', 'seller-sku', 'price', 'open-date', 'product-id', 'fulfillment-channel', 'merchant-shipping-group', 'status']]
```

```python
scopes = ["https://www.googleapis.com/auth/spreadsheets", 
          "https://www.googleapis.com/auth/drive"]
creds = Credentials.from_service_account_file("/Users/ppcintern/Desktop/Keyword_Analysis/new_credential.json", scopes=scopes)
client = gspread.authorize(creds)

# Mở Google Sheet
sheet_id = "1p3JWtN9HH1HxvARmx34Tmbt2iAjoYhK9fWIHgQr4PDE"

# Mở file Google Sheet (Spreadsheet object)
spreadsheet = client.open_by_key(sheet_id)
sheet1 = client.open_by_key(sheet_id).worksheet("Listing_US")

set_with_dataframe(sheet1, data_us)
```
